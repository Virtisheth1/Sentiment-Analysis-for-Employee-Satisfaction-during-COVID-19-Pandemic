## Step 2: Text data preprocessing

Generate a function or functions to preprocess the 6 lists (i.e., 1. apple_pros, 2. apple_cons, 3. facebook_pros, 4. facebook_cons, 5. google_pros, and 6. google_cons) with the following purposes:
1. Lower case
2. Remove leading and ending spaces
2. Remove punctuations like @ : . , ?
3. Remove new line character \n and replace it with space
4. Convert all reviews to words

Explain how your functions work using Markdown cells

```python
def review_preprocess(reviews):
    '''
    1. Lower case
    2. Remove leading and ending spaces
    3. Remove punctuations like @ : . , ?
    4. Remove new line character \n and replace it with space
    5. Convert all reviews to words
    
    You can combine all stpes in one function or seperate them into multiple functions
    '''
    preprocessed_reviews =[]
    for i in reviews:
        current = i
        current = current.lower()
        current = current.strip()
        current = current.strip('@ : . , ? ; ! / 1 2 3 4 5 6 7 8 9 0 < > * & # ^ ( ) = + " _ \ {} -')
        current = current.replace('\n',' ')
        words = current.split()
        preprocessed_reviews.extend(words)
    return preprocessed_reviews

        #reviews[i] = reviews[i].lower()
        #reviews[i] = reviews[i].strip()
        #reviews[i] = reviews[i].strip('@ : . , ?')
        #reviews[i] = reviews[i].replace('\n',' ')
        #reviews_new = reviews[i].split()
      
```

### Analyze Apple, Facebook, and Google's pro review word lists

Generate a function or functions to get the

**1. positive word count**, 
**2. stop word count**, 
**3. total word count** 

from each firm's **pro** review word list.

```python
def get_pro_scores(preprocessed_reviews):
    positive_word_count = 0
    stop_word_count = 0
    total_word_count = 0
    #print(preprocessed_reviews)
    words = review_preprocess(preprocessed_reviews)
    #print(len(words))
    #print(len(positive))
    for word in words:
        for pos_word in positive:
            if word == pos_word:
                positive_word_count +=1
        
        for stop_word in stopwords:
            if word == stop_word:
                stop_word_count +=1
    #print(positive_word_count)
    #print(stop_word_count)    
    total_word_count = len(words)
    if total_word_count - stop_word_count > 0:
        positive_score = float(positive_word_count / (total_word_count - stop_word_count))
    else:
        positive_score = 0.0
    #print( "This is total_word_count %d" % total_word_count)
    #positive_score = positive_word_count/(total_word_count - stop_word_count)


            
    '''
    Generate a function or functions to get 
    1. positive word count
    2. stop word count
    3. total word count
    from each firm's pro review word list
    
    You can combine all stpes in one function or seperate them into multiple functions
    '''

    return positive_word_count, stop_word_count, total_word_count, positive_score
```
